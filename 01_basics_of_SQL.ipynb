### What is Database ?
A Database is an organized collection of structured information, or data, typically stored electronically in a computer system.
Think of it as a digital filing cabinet. While a spreadsheet (like Excel) is great for one person to calculate numbers, a database 
is designed to hold massive amounts of data that many users can access and update at the same time securely.

### What is SQL?
SQL (Structured Query Language) is a standard language used to store, retrieve, and analyze data stored in relational databases.
In industry, SQL is used by:
- Data Analysts
- Business Analysts
- Data Scientists
- Analytics Engineers
- Backend Engineers   
Almost every company that works with data uses SQL in some form.

### Why SQL is Important for Data Analysis

SQL allows analysts to:
- Read large datasets efficiently
- Filter meaningful data
- Aggregate metrics (sales, revenue, users)
- Join data from multiple tables
- Answer business questions directly from databases   
Unlike Excel, SQL can handle millions of records reliably.

### Create Table customers

```python
%%sql 
# Drop table if it already exists (safe practice)
DROP TABLE IF EXISTS customers;
# Create customers table
CREATE TABLE customers (
customer_id INTEGER PRIMARY KEY,
name VARCHAR(50),
city VARCHAR(50),
age INTEGER,
signup_date DATE
);
```
