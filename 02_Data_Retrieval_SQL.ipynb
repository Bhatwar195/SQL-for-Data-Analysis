## ORDER BY – Sorting Data in SQL

The `ORDER BY` clause is used to **sort query results** based on one or more columns.

By default, data is sorted in **ascending order (ASC)**.
Descending order can be specified using **DESC**.

Sorting is commonly used in data analysis to:
- Rank users or products
- Find top or bottom records
- Prepare ordered reports and dashboards

### Display customer name, city, and signup date, sorted by most recent signup first.

```python
%%sql 
SELECT name,city,signup_date FROM customers
order by signup_date desc;
```

### Display customer name and age, sorted by age in descending order (oldest first).

```python
%%sql
SELECT name,age FROM customers
ORDER BY age DESC;
```

### Sort customers:
1.First by city (A–Z)  
2.Then by age (youngest to oldest) within each city

```python
%%sql
select * from customers 
order by city asc, age asc ;
```

## LIMIT – Restricting Query Results

The `LIMIT` clause is used to **restrict the number of rows** returned by a query.

It is commonly used to:
- View sample data
- Show top records
- Improve query readability

`LIMIT` is usually combined with `ORDER BY`.

### Display only the first 3 customers from the customers table

```python
%%sql 
select * from customers
limit 3 ;
```

### Display the top 3 oldest customers with their name and age.

```python
%%sql
select * from customers 
order by age desc limit 3 ;
```

### Display the 2 most recently signed-up customers with their name and signup date.

```python
%%sql 
select name ,signup_date from customers 
order by signup_date limit 3 ;
```

### Display the first 3 customers alphabetically by name.

```python
%%sql 
select name from customers 
order by name asc limit 3;
```

## WHERE – Filtering Rows in SQL

The `WHERE` clause is used to **filter rows** based on one or more conditions.

It allows you to:
- Select only relevant data
- Reduce dataset size
- Apply business rules to queries

Without `WHERE`, SQL returns **all rows** from a table.

### Display all customers who live in Mumbai.

```python
%%sql 
select * from customers
where city='Mumbai' ;
```
